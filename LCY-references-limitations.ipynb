### What analyses have already been performed on this data (or another similar dataset)? 

  





Previous research has shown that online political ads can be studied to better understand campaign messaging and targeting. One relevant study is "Analyzing Online Political Advertisements" by Sánchez Villegas, Mokaram, and Aletras (2021), which used computational methods to analyze online political advertisements. Their research shows that political ads contain important language patterns that can be studied through text analysis. This supports our project because we also use text-based methods, such as TF-IDF and buzzword analysis, to study how political ad language changes over time. However, while their study focuses more broadly on analyzing political ad content, our project focuses specifically on whether buzzwords stay stable from year to year and whether that trend differs across demographic groups. Another important study is “Systematic discrepancies in the delivery of political ads on Facebook and Instagram” by  Bär et al. (2024), which examined how political ads are delivered on Meta platforms. Their research found that the demographic groups who actually receive political ads may not always perfectly match the groups advertisers seem to target, because Meta’s delivery system also affects who sees each ad. This is important for our project because we analyze demographic patterns in political ads. Their work suggests that differences in demographic reach should be interpreted carefully, since they may reflect both campaign strategy and platform algorithms. In addition, the Wesleyan Media Project shows that political advertising datasets have been used in other academic research to study campaign communication. Altogether, these studies show that political ad data is a useful source for understanding how campaigns communicate with voters, and they provide a strong foundation for our analysis of buzzwords over time.

### What are some potential limitations and shortcomings of our analyses?



There are several limitations to our analysis. First, our methods mainly focus on TF-IDF, word frequency, and buzzword overlap, so we are not fully capturing the actual meaning of the ads. Different ads might use different words to talk about the same issue, and the same word can mean different things depending on the context. Also, some words that show up as important may just be campaign slogans, brand names, abbreviations, or slang rather than major political themes. Because of this, our results may reflect changes in word choice more than deeper changes in political messaging.

Second, our dataset was filtered before analysis, including restrictions on audience size and the removal of ads with incomplete demographic information. While this helps make the data cleaner, it also means that some ads were left out. Those excluded ads may have covered different topics or targeted different groups, so our final sample may not fully represent all political ads on the platform.

Third, even though our dataset includes around 75,000 ads across multiple years, it still may not capture every important buzzword trend or political issue over time. Some topics that were very important in real life may not appear strongly in our sample because of random chance, incomplete coverage, or the way the data was filtered. This means our findings should be interpreted as patterns in our sample, not as a complete picture of all political advertising.

Fourth, our demographic analysis is based on average audience shares, which does not directly prove advertiser intent. When we say that a buzzword is associated with a certain age group or gender group, we mean that ads containing that word tend to have audience distributions that lean toward that group. This is not the same as proving that advertisers specifically targeted only that demographic.

Fifth, our language analysis is limited because the dataset is overwhelmingly made up of English-language ads. Spanish and other-language ads make up only a small part of the sample, so comparisons across languages may be less reliable and may be strongly affected by the smaller sample sizes.

Finally, our analysis is descriptive rather than causal. We can show that buzzwords change over time and differ across groups, but we cannot say for certain why those changes happened or whether those ads were more effective.
