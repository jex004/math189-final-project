```python
import pandas as pd
```

```python
# load raw ads
df = pd.read_csv('data/raw_ads.csv')
df.head()
```

```python
# drop rows of removed content (cause you can't see the words)
removed_string = "This content was removed because it didn't follow our Advertising Standards."
df = df[~(df == removed_string).sum(axis=1).astype(bool)].reset_index(drop=True)
```

```python
# drop duplicated rows (by ad id)
df = df[~df['id'].duplicated()]
```

```python
print(df.dtypes)
df
```

```python
# correcting dataframe types
df['ad_creation_time'] = pd.to_datetime(df['ad_creation_time'], format="%Y-%m-%d")
df['ad_delivery_start_time'] = pd.to_datetime(df['ad_delivery_start_time'], format="%Y-%m-%d")
```

```python
def get_values(value):
    if not isinstance(value, str):
      return set()
    return {l.strip() for l in value.split('|')}
```

```python
# parse languages
def split_languages(s):
  en = s.map(lambda x: 'en' in get_values(x))
  es = s.map(lambda x: 'es' in get_values(x))
  other = s.map(lambda x: bool(get_values(x) - {'en', 'es'}))

  en.name, es.name, other.name = 'language_en', 'language_es', 'language_other'
  return en, es, other

df['language_en'], df['language_es'], df['language_other'] = split_languages(df['languages'])
```

```python
# parse publisher_platforms column
def split_publisher_platforms(s):
  fb = s.map(lambda x: 'facebook' in get_values(x))
  ig = s.map(lambda x: 'instagram' in get_values(x))
  other = s.map(lambda x: bool(get_values(x) - {'facebook', 'instagram'}))
  
  fb.name, ig.name, other.name = 'platform_fb', 'platform_ig', 'platform_other'
  return fb, ig, other

df['platform_fb'], df['platform_ig'], df['platform_other'] = split_publisher_platforms(df['publisher_platforms'])
```

```python
# parse demographic_distribution column
def expand_demographics(s):
  def parse_entry(value):
    if not isinstance(value, str):
      return {}
    result = {}
    for part in value.split('|'):
      part = part.strip()
      if not part:
        continue
      key, pct = part.split(':')
      result[key.strip()] = float(pct.strip().replace('%', ''))
    return result

  parsed = s.map(parse_entry)
  return pd.DataFrame(parsed.tolist(), index=s.index)
  
demographics_df = expand_demographics(df['demographic_distribution'])

# data cleaning to demographics
funny_column_name = 'All (Automated App Ads)/All (Automated App Ads)'
demographics_df = demographics_df.drop(columns=funny_column_name).fillna(float(0))
demographics_df = demographics_df.reindex(sorted(demographics_df.columns), axis=1)

# add to original df
df = pd.concat([df, demographics_df], axis=1)
```

```python
# drop unnecessary columns
df = df.drop(columns="audience_size_upper") # too many nan values
df = df.drop(columns="delivery_by_region") # this column sucks, if we need it i'll bring it back
df = df.drop(columns=["languages", "publisher_platforms", "demographic_distribution"]) # already parsed
```
