## Cosine Similarity Test


### Results
The data shows that political buzzwords have stayed very consistent from 2019 to 2025. By using a permutation test, the overlap in words between different years is significantly higher than what would happen by random chance.

### Interpretation
Political language is very stable and does not change much from year to year.

### Potential Explanations
Campaigns stick to a specific set of words they know work for fundraising and getting attention, such as action, donate, tax, and freedom. The only time this pattern really broke was between 2020 and 2024, which is likely because the specific language used during the pandemic did not apply to the 2024 election cycle or climate which focused more heavily on topics like abortion, reproductive rights, and insurance.

## Chi-Squared Test for Homogeneity (Age)


### Results
The Chi-Square results for age groups show significant p-values for most of the top buzzwords, leading tot he rejection of the null hypothesis.

### Interpretation
Political ads are being micro-targeted based on age.

### Potential Explanations
Advertisers pick specific words based on how old the person is because they want to focus on issues relevant to that stage of life. Campaigns might use words like debt or future for younger people while switching to benefits or seniors for an older audience to make the ads more effective.

## Chi-Squared Test for Homogeneity (Gender)


### Results
The test for gender showed a significant difference in word frequency, leading to the rejection of the null hypothesis.

### Interpretation
Political messaging is gendered suggesting that campaigns talk to men and women differently.

### Potential Explanations
Advertisers use specific hooks to grab attention based on gender. They likely assume that certain topics like healthcare or reproductive rights will engage women, while different terms like veterans and disability are pushed towards men. This shows that campaigns change their language to match the gender of the person looking at the ad instead of sending one general message to everyone.
